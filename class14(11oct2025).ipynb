```python
cities = ["Islamabad" , "Lahore" , "Karachi" , "Peshawar" , "Multan"]
count = 0 
vowels = "AEIOUaeious"
resultList = []
for city in cities:
    newCity = ""
    for ch in city:
        if ch not in vowels:
            newCity+=ch
    resultList.append(newCity)
        
print(resultList)
```

```python
cities = ["Islamabad", "Lahore", "Karachi", "Peshawar", "Multan", "Quetta"]
unique_cities = []

for city in cities:
    city_lower = city.lower()  # lowercase for fair checking
    if len(set(city_lower)) == len(city_lower):  # if all chars are unique
        unique_cities.append(city)

print("Cities with all unique characters:", unique_cities)
```
